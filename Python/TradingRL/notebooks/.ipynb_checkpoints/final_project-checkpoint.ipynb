### Get data

### Model 1: Simple Linear regression

This model assumes that we are (linearly) regressing the price against the lagged price (in fact the logarithms). Therefore this model would be reasonable as long as the scatter plot gold_t vs gold_{t-1} is linear. Let us check it.

```python
plt.figure()
plt.plot(df['lagged'], df['logUSD'], 'bo')
plt.grid()
plt.title("Lagged vs current price 1978 until 2014") 
plt.xlabel("log lagged price")
plt.ylabel("log price")
plt.grid(True)
plt.show()
```

We now perform a linear regression

```python
# linear regression
gold = df['logUSD'][1:]
lag = df['lagged'][1:]
    
beta, alpha, r_value, p_value, std_err = stats.linregress(gold, lag)

print beta, alpha # the estimated parameters 0.926627373031 0.415535989119
print "R-squared=", r_value**2 #R-squared= 0.860683037219
print "p-value =", p_value #p-value = 5.54104692809e-17
```

The coefficient of determination shows that gold price in the previous period explains 86% of today's evolution we have obtained a fairly reasonable fit but are we happy with a 86% prediction? If volatility is high then this can be a bad model.

### Model 2: ARIMA(p,q)

We now apply an ARIMA model. First we test is there is unit root in out series.

```python
# Test unit-roots:
adf_res = stattools.adfuller(gold, maxlag=2, regression='c', autolag='AIC', store=False, regresults=False)
p_value = adf_res[1]
print "Log-price p-value: ", p_value
    
if p_value > 0.05:
    print "We cannot reject the null hypothesis: there is a unit-root"
else:
    print "We reject the null hypothesis: the process is stationary"
```

We can see that there are indications that the series is non-stationary. Next we look at the ACF and PACF for indications about the arma orders.

```python
# plot the acf() and the pacf()
tsaplots.plot_acf(gold, lags=20);
tsaplots.plot_pacf(gold, lags=10);
```

The sinusoidal form of the ACF and the PCF suggests some AR(2). We will try different models and select the one with lowest AIC. The following function perform safe arima fits.

```python
def safe_arima_fit(y, order, trend, method, start_params=None):
    
    """
    Performs ARIMA fit

    Parameters
    ----------
    y : array-like
        Time-series data
    order : iterable
        The (p,d,q) order of the model for the number of AR parameters,
        differences, and MA parameters to use.
    trend : str
        The trend to use when fitting the ARIMA models.
        'c': constant trend; 'nc': no trend
    model_kw : dict
        Keyword arguments to be passed to the ``ARIMA`` model
    fit_kw : dict
        Keyword arguments to be passed to ``ARIMA.fit``.

    Returns
    -------
    obj : Results object
    """
    
    try:
        
        model = arima_model.ARIMA(y, order=order)
        return model.fit(disp=0, trend=trend,
                         start_params=start_params, method=method)

    except LinAlgError:
        # SVD convergence failure on badly misspecified models
        return

    except ValueError as error:
        if start_params is not None:  # don't recurse again
            # user supplied start_params only get one chance
            return
        # try a little harder, should be handled in fit really
        elif ('initial' not in error.args[0] or 'initial' in str(error)):
            start_params = [.1] * sum(order)
            if trend == 'c':
                start_params = [.1] + start_params
            return safe_arima_fit(y, order, trend, method, start_params)
        else:
            return
    except:  # no idea what happened
        return
```

```python
# Automatic ARIMA model selection:
order_list = [(1,0,0), (2,0,0), (2,0,0), # AR
              (1,1,0), (2,1,0), (3,1,0), # AR + differencing
              (2,1,1), (3,1,1), # ARIMA(p,1,1)
              (2,1,2), (3,1,2), # ARIMA(p,1,2)
              (2,1,3), (3,1,3)] # ARIMA(p,1,3)

aic_df = pd.DataFrame(np.nan, index=range(len(order_list)), columns = ['aic', 'i', 'j', 'k'])

for i, order in zip(range(len(order_list)), order_list):
    
    fit = safe_arima_fit(gold.values, order, trend='nc', method='css-mle')
    
    aic_df.loc[i, 'aic'] = fit.aic
    aic_df.loc[i, 'i'] = order[0]   
    aic_df.loc[i, 'j'] = order[1]    
    aic_df.loc[i, 'k'] = order[2]               

# drop unused rows
aic_df=aic_df.dropna()
aic_df
```

The selected model is the one with lowest AIC, as can be seen in the following figure.

```python
# plot figure with different models    
plt.figure()
plt.scatter(aic_df.index, aic_df.aic)
plt.plot(aic_df.aic, '-')
plt.title('ARIMA(p,d,q) AIC evolution', fontdict=font)
plt.ylabel('AIC', fontdict=font)
plt.xlabel('Model number', fontdict=font)
plt.xlim(0,len(aic_df))
plt.tight_layout()
plt.grid(True)

xoffset = 0.2
yoffset = 1

for i in range(len(aic_df)):
    order = [int(l) for l in aic_df.iloc[i, 1:4].tolist()]
    plt.text(aic_df.index[i]+xoffset, aic_df.iloc[i, 0]+yoffset, str(order),
             color='red', fontsize = 6)

plt.show()
```

We can see that simple regression ARIMA(1,0,0) $$r_t = a + b\ r_{t-1}$$ is the worst of all possible models. However, simple differencing would have made this model the best one. Therefore we select an ARIMA(1,1,0) model.

```python
aic_df.iloc[3, :]
```

```python
order = (1,1,0)
model = arima_model.ARIMA(gold.values, order)
fit = model.fit(method="mle")
fit.summary2()
```

```python
# p-values
print(fit.pvalues)

# QQ plot
sm.qqplot(fit.resid, line='q');

# ACF of residuals
tsaplots.plot_acf(fit.resid);
```

From the QQ plot we can see quite a bit of effect from fat tails. Also, we check the ACF of the residuals and verify that we get a zero-lag (white noise process), therefore we have achieved a good fit with ARIMA(1,1,0).

Now, is it necessary to model volatility. We need to check for evidence of conditional heteroskedastic (squared-residuals). To do so we plot the ACF of the squared-residuals.

```python
tsaplots.plot_acf(fit.resid**2);
```

The plot shows no evidence of serial correlation in the squared residuals, then no GARCH needed.

According to the ARIMA(1,1,0) model the forecasted Gold price is the following:

```python
# one-step out of sample forecast
start = gold.index.get_loc(2016)
end = start + 1
forecast = fit.predict(start=start, end=end, typ='levels', dynamic=False)
    
print "forecast in 2016 (ARIMA(1,1,0)):", np.exp(forecast[0])
print "forecast in 2017 (ARIMA(1,1,0)):", np.exp(forecast[1])
```

### Model 3: VAR

We now implement a VAR model using specific factors. VAR assumes that the time series are stationary, so we take differences to eliminate possible unit-root processes.

```python
# select factors and make series stationary
factors = ['LogGold_Lag'
           ,'PPIACO'
           ,'CPIAUCNS'
           ,'M1NS'
           ,'UNRATE'
           ,'INDPRO'
           #,'NASDAQ'
           ,'SP500'
           ,'TB3MS'
           #,'TNX'
           #,'TWEXBPA'
          ]

mdata = data[['LogGold'] + factors].diff().dropna()
#    mdata = data.diff().dropna()

# make a VAR model
model = VAR(mdata)
    
# select appropriate lag in the model
model.select_order(maxlags=5)

# fit the VAR model 
results = model.fit(3)
print(results.summary())
```

```python
# plot autocorrelations
results.plot_acorr()
```

We now study the effect of the selected factor in gold price changes by looking at the impulse response functions 

```python
# Impulse responses
irf = results.irf(3)
irf.plot(impulse='LogGold')
```

We can see that IP series is also non-stationary, which is also patent from the ACF plot.

```python
# impulse response function (cumulative effects)
irf.plot_cum_effects(impulse='LogGold')
```

We also check for Ganger's causality for selected factors:

```python
# check for causality
results.test_causality('LogGold', factors, kind='f')
```

We cannot conclude that this group of selecter factors Ganger cause changes in gold prices. Finaly we perform a dynamic forecast using the VAR model.

```python
# Dynamic forecast
var = DynamicVAR(data, lag_order=2, window_type='expanding')
var.forecast(2)
var.plot_forecast(2)
```

### Model 4:  Multi-linear regression & Support Vector Regression (SVR)

We now introduce regression models, specifically multi-linear and support vectors. We study how these model perform using backtesting.

```python
X = data.values[:,1:] # numpy array or sparse matrix of shape [n_samples,n_features]
y = data.values[:,0] # numpy array of shape [n_samples, n_targets]

# - Split data into random train and test subsets using train_test_split
# this method split the data randomly, and therefore 
# it cannot be directly used with time series data
#    X_train, X_test, y_train, y_test = train_test_split(X, y)

# - Manual train-test split ( 66% - 34%)
train_size = int(len(X) * 0.66)
X_train, X_test = X[0:train_size, :], X[train_size:len(X), :]
y_train, y_test = y[0:train_size], y[train_size:len(y)]

print('Observations: %d' % (len(X)))
print('Training Observations: %d' % (len(X_train)))
print('Testing Observations: %d' % (len(X_test)))
```

#### Multilinear Regression

```python
lm = LinearRegression()
res_lm = lm.fit(X_train, y_train)
pred_lm = res_lm.predict(X)
print("Score: ", res_lm.score(X_test, y_test))
print("Regression coefficients: ", res_lm.coef_)
```

#### Support Vector Regression

```python
svr = SVR(kernel='linear')
res_svr = svr.fit(X_train,y_train)
pred_SVR = res_svr.predict(X)
print("Score: ", res_svr.score(X_test, y_test))
print("Regression coefficients: ", res_svr.coef_)
```

#### Backtesting

```python
plt.plot(y_train, color='darkorange', marker='o', label='Train data', alpha=0.5)
plt.plot([None for i in y_train] + [x for x in y_test], color='blue', marker='o', label='Test data', alpha=0.5)
plt.plot(pred_SVR, color='green', marker='o', label='SVR', alpha=0.25)
plt.plot(pred_lm, color='red', marker='o', label='LM', alpha=0.25)
plt.ylabel('Log Gold Price (target)')
plt.xlabel('Time index')
plt.title('Factor Models Prediction')
plt.legend()
plt.show()
```

As we can see, multilinear regression apparently does a better job in predicting price, but both are able to anticipate the drop at the end of the series.
