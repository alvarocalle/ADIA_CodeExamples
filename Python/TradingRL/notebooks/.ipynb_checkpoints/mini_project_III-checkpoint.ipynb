## Augmented Dickey-Fuller (ADF) Test

Let us apply the Augmented Dickey-Fuller (ADF) Test to see if data presents a unit root. We use adfuller function from Statsmodels module in Python:

- Null hypothesis: There is a unit root (gamma = 0)
- Alternative hypothesis: There is no unit root

Let us first apply the test to the price series

```python
hpi = df.CSUSHPINSA

# ADF Test of the prices:
adf_res = stattools.adfuller(hpi, maxlag=2, regression='c', autolag='AIC', store=False, regresults=False)
p_value = adf_res[1]
print "Price p-value: ", p_value

if p_value > 0.05:
    print "We cannot reject the null hypothesis: there is a unit-root"
else:
    print "We reject the null hypothesis: the process is stationary"\
```

We now take logarithms of prices

```python
lhpi = np.log(hpi)
lhpi.plot()
```

We see that although we now have an upward linear trend, the variance stay constant.

```python
# ADF Test of the logarithms:
adf_res = stattools.adfuller(lhpi, maxlag=2, regression='c', autolag='AIC', store=False, regresults=False)
p_value = adf_res[1]
print "Log-price p-value: ", p_value
```

**Comment:**

The results provided pvalue = 0.964 which is the higher as compared to the established level of 0.05. Therefore we do not reject the null hypothesis.

- Inflation data presents a unit root and therefore it is not stationary.
- We see that taking logs makes the process almost stationary.
- We obtain stationarity by applying the first difference

## Implementation of the ARIMA(p,d,q) model

We now determine the (p, d, q) ARIMA orders using the Box-Jenkins methodology. Some rules of thumb are the following:

- AR models have theoretical PACFs with non-zero values at the AR terms in the model and zero values elsewhere. The ACF will taper to zero in some fashion.
- An AR(2) has a sinusoidal ACF that converges to 0.
- MA models have theoretical ACFs with non-zero values at the MA terms in the model and zero values elsewhere.
- ARMA models (including both AR and MA terms) have ACFs and PACFs that both tail off to 0.  These are the trickiest because the order will not be particularly obvious. 
- If the ACF and PACF do not tail off, but instead have values that stay close to 1 over many lags, the series is non-stationary and differencing will be needed.  Try a first difference and then look at the ACF and PACF of the differenced data.
- If all autocorrelations are non-significant, then the series is random (white noise). We are done at that point.
- If after having taken first differences, all autocorrelations are non-significant, then the series is called a random walk and we are done.

```python
# We first plot the acf() and the pacf() of the logarithmic series
tsaplots.plot_acf(lhpi, lags=10);
tsaplots.plot_pacf(lhpi, lags=10);
```

Observations:

1. As we can see the ACF does not go down to zero in any reasonable amount of time and we confirm the non-stationarity of the series.
2. However, the PACF falls to zero after one lag which suggests that some AR(1) process is going on, at the very least.

Lets remove the trend by differencing, and replot the ACF and PACF.

```python
dhpi=lhpi.diff()[1:]
dhpi.plot()
        
adf_res = stattools.adfuller(dhpi, maxlag=2, regression='c', autolag='AIC', store=False, regresults=False)
p_value = adf_res[1]

print "Log-diff price p-value: ", p_value
```

Now the time series is clearly stationary according to the ADF test

```python
# plot acf() and pacf() after differencing
tsaplots.plot_acf(dhpi, lags=10);   
tsaplots.plot_pacf(dhpi, lags=20);
```

The plots above suggest that we will have a ARIMA(p,d,q) probably with $p \geq 1$, $d=1$ and $q \geq 0$. 

As a summary of ACF, PACF and order estimation:

MODEL | ACF | PACF  
 -----|-----|------
AR(p)| Consists of damped exponential or sine waves, dies out exponentiall | Is zero after p lags (can be spike at p)
MA(q)| Is zero after q lags (can be spike at q) | Consists of mixtures of damped exponential or sine terms, dies out exponentially
ARMA(p,q)| Eventually dominated by AR(p) part, then dies out exponentially | Eventually dominated by MA(q) part, then dies out

We now proceed with the fit of the ARIMA model. We first create the following function that handles convergence problems. The function returns an ARIMA fit results as far as the solution has converged.

```python
from numpy.linalg import LinAlgError

def safe_arima_fit(y, order, trend, method, start_params=None):
    
    """
    Performs ARIMA fit

    Parameters
    ----------
    y : array-like
        Time-series data
    order : iterable
        The (p,d,q) order of the model for the number of AR parameters,
        differences, and MA parameters to use.
    trend : str
        The trend to use when fitting the ARIMA models.
        'c': constant trend; 'nc': no trend
    model_kw : dict
        Keyword arguments to be passed to the ``ARIMA`` model
    fit_kw : dict
        Keyword arguments to be passed to ``ARIMA.fit``.

    Returns
    -------
    obj : Results object
    """
    
    try:
        
        model = arima_model.ARIMA(y, order=order)
        return model.fit(disp=0, trend=trend,
                         start_params=start_params, method=method)

    except LinAlgError:
        # SVD convergence failure on badly misspecified models
        return

    except ValueError as error:
        if start_params is not None:  # don't recurse again
            # user supplied start_params only get one chance
            return
        # try a little harder, should be handled in fit really
        elif ('initial' not in error.args[0] or 'initial' in str(error)):
            start_params = [.1] * sum(order)
            if trend == 'c':
                start_params = [.1] + start_params
            return safe_arma_fit(y, order, trend, method, start_params)
        else:
            return
    except:  # no idea what happened
        return
```

Now we fit different models:

```python
# ARIMA model selection:
aic_df = pd.DataFrame(np.nan, index=range(18), columns = ['aic', 'i', 'j', 'k'])

order_list = [(1,0,0), (2,0,0), # AR
              (1,1,0), (2,1,0), (3,1,0), (4,1,0), (5,1,0), # AR + differencing
              (1,1,1), (2,1,1), (3,1,1), (5,1,1), # ARIMA(p,1,1)
              #(1,1,2), (2,1,2), (4,1,2), # ARIMA(p,1,2)
              #(1,1,3), (2,1,3), (4,1,3), (5,1,3)
             ] # ARIMA(p,1,3)

for i, order in zip(range(18), order_list):
    
    fit = safe_arima_fit(lhpi, order, trend='nc', method='css-mle')
    
    aic_df.loc[i, 'aic'] = fit.aic
    aic_df.loc[i, 'i'] = order[0]   
    aic_df.loc[i, 'j'] = order[1]    
    aic_df.loc[i, 'k'] = order[2]
    
# drop unused rows
aic_df=aic_df.dropna()
```

## Forecasting

We now forcast the future evolution of Case-Shiller Index using the obtained ARIMA model.

- Provide one month, two-month and three-month forecasts.
- Test the model using in-sample forecasts. 
- Suggest exogenous variables that can be introduced which can improve the forecasts. 

Lets look at our in-sample prediction. We will use our fitted model which has been selected using the whole sample to predict future values. We start our prediction from 2001-01-01 and go out-of-the-sample.

```python
# one-step out of sample forecast
start=lhpi.index.get_loc(datetime.datetime(2001, 1, 1))
end = start + 200
forecast = fit.predict(start=start, end=end, typ='levels', dynamic=False)
    
plt.figure()
plt.plot(lhpi, 'bo', alpha= 0.2, label="Data" )
plt.plot(forecast, 'r-', label="Prediction")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title('Case-Shiller U.S. National Home Price Index', fontdict=font)
plt.ylabel('Log-Prices', fontdict=font)
plt.xticks(rotation=45)
plt.show()
```

Prices are obtained taking the exponential of the logarithms.

```python
# Price prediction
plt.figure()
plt.plot(hpi, 'bo', alpha= 0.2, label="Data" )
plt.plot(np.exp(forecast), 'r-', label="Prediction")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title('Case-Shiller U.S. National Home Price Index', fontdict=font)
plt.ylabel('Prices', fontdict=font)
plt.xticks(rotation=45)
plt.show()
```
